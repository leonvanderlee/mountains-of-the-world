# __Visualize__

```python
df = pd.read_csv('../data/processed/mountains.csv', )

df
```

```python
import yaml

with open(r'../config/token.yaml') as file:
    data = yaml.load(file, Loader=yaml.FullLoader)
    token = data['TOKEN']
```

```python
def scatter_mapbox(filename, size_max, zoom_level, center_lat, center_lot, color_min, color_max):
        fig = px.scatter_mapbox(
                df,
                lat='Latitude',
                lon='Longitude',
                color='Height (meters)',
                hover_name='Name',
                size='Height (meters)',
                color_continuous_scale=px.colors.sequential.Inferno,
                range_color=[color_min, color_max],
                size_max=size_max,
                zoom=zoom_level,
                center={'lat': center_lat, 'lon': center_lot},
                width=1280,
                height=640
        )

        fig.update_layout(
                margin={'t': 0, 'b': 0, 'l': 0, 'r': 0},
                mapbox={
                        'accesstoken': token,
                        'style': 'outdoors',

                },
        )

        fig.write_image(
                f'../reports/figures/{filename}.jpg',
                format='jpg',
                scale=2,
                engine='kaleido',
        )

        fig.show()
```

```python
scatter_mapbox(filename='alps', size_max=15, zoom_level=6.5, center_lat=46, center_lot=9, color_min=2000, color_max=4800)
```
